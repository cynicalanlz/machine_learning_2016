```python
def check_best_scores_equalness(best_scores):
    best_scores_equalness = np.equal(best_scores[0], best_scores)
    t = True
    for q in best_scores_equalness:
        t = np.logical_and(t,q)
    return t

check_best_scores_equalness(best_scores)
```

```python
def get_average_scores(classes, features,n, folds, rnd):    
    
    sc_ = []
    
    kf = KFold(len(features), n_folds=folds, shuffle=True, random_state=rnd)
    
    for k in list(xrange(n+1))[1:]:        
        neigh = KNeighborsClassifier(n_neighbors=k)
        neigh.fit(features, classes) 
        scores = cross_val_score(neigh, features, classes, cv=kf)
        sc_.append(scores)    
    
    sc_avg = map(np.average,sc_)
    ans_ = dict(zip(xrange(len(sc_avg)),sc_avg))
    extrem_ = [x for x in map(lambda x: (x[0]==max(sc_avg))*(x[1]+1), zip(ans_.values(), xrange(len(ans_.values())))) if x!=0]
    extrem_ = map(int,extrem_-np.ones(len(extrem_)))    
    extrem = max(sc_avg)    
    
    return sc_, ans_, np.average(sc_avg[np.argmax(sc_avg)]), extrem_
```
