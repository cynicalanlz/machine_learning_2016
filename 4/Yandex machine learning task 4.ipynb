```python
import pandas
import numpy as np
from sklearn.externals.six import StringIO
from sklearn import tree
import pydot
from sklearn.datasets import load_boston
from IPython.display import Image
import copy
from sklearn.neighbors import KNeighborsRegressor
from sklearn.cross_validation import KFold
from sklearn.cross_validation import cross_val_score
from sklearn.preprocessing import scale

import operator
import itertools

def fit_KNeighborsRegressor(classes, features, test_space, n, folds, rnd):    
    
    sc_ = []
    
    kf = KFold(len(features), n_folds=folds, shuffle=True, random_state=rnd)
    
    for q in test_space:        
        neigh = KNeighborsRegressor(p=q,n_neighbors=n,weights='distance')
        neigh.fit(features, classes) 
        scores = cross_val_score(neigh, features, classes, cv=kf,scoring='mean_squared_error')
        sc_.append(scores)    
    
    sc_avg = map(np.average,sc_)
    #sc_avg = map(abs, sc_avg)
    ans_ = dict(zip(xrange(len(sc_avg)),sc_avg))
    extrem_ = [x for x in map(lambda x: (x[0]==max(sc_avg))*(x[1]+1), zip(ans_.values(), xrange(len(ans_.values())))) if x!=0]
    extrem_ = map(int,extrem_-np.ones(len(extrem_)))    
    
    extrem = max(sc_avg)
    
    return sc_, ans_, np.average(sc_avg[np.argmax(sc_avg)]), extrem_, dict(zip(xrange(len(test_space)),test_space))

dt = load_boston()
len(dt.data)==len(dt.target), dt.data.shape
```

```python
dt.data = scale(dt.data)
test_space = np.linspace(1,10,num=200)
neighbors = 5
folds = 5
random_state = 42
```

```python
ans = fit_KNeighborsRegressor(dt.target,dt.data, test_space, neighbors,folds, random_state)
```
